## Combine Your Data

Now that you have four clean sets of data, combine all four into one dataframe that represents the entire Northeast region.

```python
northeast_df=baltimore_df.append([boston_df, newyork_df, philadelphia_df])
northeast_df.info()
```

```python
# Combine the four dataframes into one!
for col in northeast_df.columns:
    pct_missing = np.mean(northeast_df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
    
for col in northeast_df.columns:
    pct_missing = np.mean(northeast_df[col].isnull())
    if(pct_missing==1):
        northeast_df=northeast_df.drop([col], axis=1)
northeast_df.info()
```
